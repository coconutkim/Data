# 히트 맵 heat map

열을 뜻하는 히트(heat)와 지도를 뜻하는 맵(map)을 결합시킨 단어

색상으로 표현할 수 있는 다양한 정보를 일정한 이미지위에 열분포 형태의 비쥬얼한 그래픽으로 출력하는 것이 특징이다



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 예제 데이터
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [2, 4, 6, 8, 10],
    'C': [5, 4, 3, 2, 1]
}

df = pd.DataFrame(data)

# 상관계수 계산
correlation_matrix = df.corr()

# 히트맵으로 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

# 커널 밀도 추정 kde

Kernel density estimation, KDE

통계학에서 확률 밀도 추정을 위해 커널 평활화를 적용한 것

즉, 커널을 가중치로 기반으로 랜덤 변수의 확률 밀도 함수를 추정하는 비모수적 방법이다



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 예제 데이터 생성
np.random.seed(42)
data = pd.DataFrame({
    'Column1': np.random.normal(0, 1, 1000),
    'Column2': np.random.normal(5, 2, 1000),
    'Column3': np.random.normal(-2, 1.5, 1000),
    'Column4': np.random.normal(3, 0.5, 1000)
})

# 전체 데이터 KDE 그래프 (초록색)
sns.kdeplot(data.values.flatten(), color='green', label='Overall KDE')

# 특정 3개의 칼럼에 대해 KDE 그래프 (파란색)
sns.kdeplot(data['Column1'], color='blue', label='Column1')
sns.kdeplot(data['Column2'], color='blue', label='Column2')
sns.kdeplot(data['Column3'], color='blue', label='Column3')

# 그래프 설정
plt.legend()
plt.title('Kernel Density Estimation (KDE) Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
