## ravel(): 그래프를 그리기 위해서 왜 데이터가 1차원이여야 하는가

2D 그래프의 본질:

2D 그래프는 보통 x축과 y축으로 구성됩니다. x축에는 독립 변수, y축에는 종속 변수가 위치합니다.
이 경우, 각 축에 대한 데이터는 1차원 배열이나 리스트로 표현되며, 각 배열의 원소들이 각각의 축의 값을 나타냅니다.
데이터의 대응 관계:

그래프를 그릴 때, 독립 변수(x축)와 종속 변수(y축) 사이의 대응 관계를 보여주는 것이 일반적입니다. 이 관계는 데이터가 1차원 배열로 제공될 때 가장 잘 나타낼 수 있습니다.
예를 들어, x축이 시간(time)이고 y축이 온도(temperature)라면, 시간이 1차원 배열로, 온도도 1차원 배열로 제공되며, 각 원소는 특정 시간에 해당하는 온도를 나타냅니다.
복잡성을 줄임:

데이터를 1차원으로 평탄화하는 것은 그래프를 그릴 때 불필요한 복잡성을 줄입니다. 다차원 데이터를 평탄화하면 데이터 포인트를 쉽게 매핑하고 시각화할 수 있습니다.
그래프 라이브러리의 요구사항:

많은 그래프 라이브러리에서 데이터가 1차원 형태로 제공되기를 요구합니다. 예를 들어, matplotlib나 seaborn 같은 라이브러리에서는 x와 y의 데이터가 각각 1차원 배열로 주어져야 각 데이터 포인트를 올바르게 매핑하고 시각화할 수 있습니다.

```python
plt.figure(figsize=(10,6))
sns.histplot(df2,bins=np.arange(50,100,5),kde=True)
sns.histplot(df2_after,bins=np.arange(50,100,5),kde=True) #arange(최소,최대,구간)
plt.show()
```
