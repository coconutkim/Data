># 모집단과 표본

> 모집단 population

통계분석 방법을 적용할 관심 대상의 전체 집합

> 표본 sample

과학적인 절차를 적용하여 모집단을 대표할 수 있는 일부를 추출하여 직접적인 조사 대상이 된 모집단의 일부

>모수 parameter 

모집단을 분석하여 얻어지는 결과 수치

>통계량 statistic

표본을 분석하여 얻어지는 결과 수치



μ (뮤): 모평균

σ (시그마): 모표준편차

σ² (시그마 제곱): 모분산

Σ (시그마): 합계

π (파이): 확률을 나타낼 때 사용될 수 있으며, 이항분포에서 성공 확률을 나타내기도 합니다

α (알파): 유의수준을 나타냅니다. 보통 0.05나 0.01과 같은 값을 사용하며, 귀무가설을 기각할 임계값을 나타냅니다

β (베타): 제2종 오류의 확률 또는 회귀분석에서 독립변수의 계수를 나타냅니다

θ (세타): 모수를 나타내는 데 사용됩니다. 예를 들어, 모평균이나 모표준편차와 같은 특정한 모집단의 특성을 나타냅니다

λ (람다): 포아송 분포의 평균이나, 특정 확률 분포의 모수를 나타냅니다

ρ (로): 모집단 상관계수

η (에타): 주로 효과 크기(effect size)를 나타내기 위해 사용됩니다. 특히, 분산분석(ANOVA)에서 사용되는 **η² (에타 제곱)**는 독립변수가 종속변수의 총 분산 중에서 설명할 수 있는 비율을 나타냅니다

># 데이터의 구분

> 범주형 척도 categorical scale

데이터들을 구분지어 나눌 수 있는 척도

>> 명목척도 nominal scale

수(數) 또는 순서의 개념과는 상관없이 이름만 붙여지는 척도

>> 서열척도 orginal scale

명목척도와 유사하게 숫자 혹은 연산과는 연관이 없지만, 순서(서열)을 구분할 수 있는 척도

> 연속형 척도 continuous scale 

연결된 속성의 데이터를 조사의 목적에 맞게 구분한 척도

>> 등간척도 interval scale 

명목척도나 서열척도와 달리, 측정된 자료들 간에 
더하기와 빼기가 가능한 척도

>> 비율척도 ratio scale

등간척도의 성질과 함께 무(無)의 개념인 0값도 가지는 척도

#### 데이터의 구분 예시

>명목

성별: 남성, 여성, 비바이너리

혈액형: A형, B형, AB형, O형

도시: 서울, 뉴욕, 런던

>서열

교육 수준: 고등학교 졸업, 대학 재학 중, 대학교 졸업

만족도 평가: 매우 불만족, 불만족, 보통, 만족, 매우 만족

랭킹: 1위, 2위, 3위

>등간

온도: 섭씨 온도(예: 20°C, 25°C) — 절대적인 0도가 없으므로 온도의 간격은 의미가 있지만, 0°C는 "온도가 없음"을 의미하지 않음

날짜: 날짜와 시간 (예: 2024년 8월 11일) — 특정 기준점이 없고, 간격은 의미가 있음

>비율

길이: 미터(m), 센티미터(cm) — 0미터는 "길이 없음"을 의미하며, 비교와 비율 계산이 가능함

무게: 킬로그램(kg), 그램(g) — 0킬로그램은 "무게 없음"을 의미하며, 비율 계산이 가능함

나이: 년 (예: 25세) — 0세는 "나이 없음"을 의미하며, 나이를 비교하고 비율을 계산할 수 있음



># 산포도 dispersion

중심경향도만으로는 집단에 대한 성격과 분포를 파악하는데 부족하다

측정된 데이터가 어떻게 분포하고 있는지에 대해 파악해야 한다

데이터를 제대로 이해하기 위하여 표본이 가지는 분포의 정도를 나타내는 산포도(dispersion)를 확인해야 한다
