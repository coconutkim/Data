## 뉴스 기사 가져오기

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
code = requests.get('https://n.news.naver.com/article/014/0005222896')
```

```python
code. raise_for_status()
```

```python
code.content
```

```python
type(code.content)
```

```python
bs = BeautifulSoup(code.content,'html.parser')
```

```python
target = 'newsct_article _article_body'
```

```python
bss = bs.find_all(class_ = target)
```

```python
for element in bss:
    print(element.get_text())
```

```python
with open('estateNews.txt','w', encoding='utf-8') as file:
    for element in bs:
        file.write(element.get_text())
```
