## 연릭방정식
연립방정식이란 미지수가 여러개 포함된 방정식을 묶어 놓은 것을 의미한다.
아래는 연립방정식이다

$3x+y=2$

$x-2y=3$

```python
x = Symbol('x')
y = Symbol('y')
```

```python
equation1 = 3*x+y-2
equation2 = x-2*y-3
```

```python
solve((equation1,equation2)) #연립방정식 풀 때는 튜플을 이용한다
```
