mosaic chart



```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
```

```python
# 데이터 생성
data = sns.load_dataset('tips')

# 모자이크 그래프 그리기
mosaic(data=data, index=['sex', 'smoker'], title='Mosaic Plot of Gender and Smoking Status')

# 그래프 출력
plt.show()
```
