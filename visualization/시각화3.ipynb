> linspace

start: 생성할 수열의 시작 값.

stop: 생성할 수열의 끝 값.

num: 생성할 샘플의 개수 (기본값: 50).

endpoint: True일 경우, stop 값이 포함됩니다. False일 경우, stop 값이 포함되지 않음.

retstep: True일 경우, 생성된 샘플들 간의 간격도 함께 반환합니다.

dtype: 반환되는 배열의 자료형.

axis: 샘플이 배치될 축.

```python
x = np.linspace(-4,4,100)
y1 = x**3
y2 = 10*x**2-1

plt.plot(x,y1,'o-r',x,y2,'*--b')
plt.xlim(-1,1) #범위를 설정하여 확대한다
plt.ylim(-3,3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('my plot')
plt.legend(['$y1=x^3$','$y2=10x^2-1$'],loc = 'upper right')
plt.show()
```

>plt.legend

Matplotlib 라이브러리에서 그래프에 범례(legend)를 추가하는 데 사용되는 함수

범례는 그래프에 표시된 데이터 시리즈(라인, 바, 점 등)의 레이블을 나타낸다

그래프의 각 요소가 무엇을 나타내는지 이해하는 데 도움을 줍니다

loc

범례의 위치를 지정합니다

loc='upper right', loc='lower left' 등으로 사용할 수 있으며, 기본적으로 자동 위치가 설정됩니다.

'best' (기본값): Matplotlib가 자동으로 최적의 위치를 선택

'upper right', 'upper left', 'lower right', 'lower left': 그래프의 각 모서리 위치

'center', 'center left', 'center right': 중앙 위치

'upper center', 'lower center': 상단 또는 하단 중앙

fontsize: 범례 텍스트의 글꼴 크기를 설정합니다.

title: 범례에 제목을 추가할 수 있습니다.

shadow: True로 설정하면 범례에 그림자를 추가할 수 있습니다.

ncol: 범례 항목을 여러 열로 표시할 수 있습니다.
