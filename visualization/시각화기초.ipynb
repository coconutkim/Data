### [<matplotlib.lines.Line2D at 0x23445b73350>]의 의미

Matplotlib에서 플롯(예: 선 그래프)을 그릴 때, 그래프를 시각적으로 표시하지 않고, 단순히 해당 객체를 출력할 때 나타나는 Python 객체의 표현입니다

```python
plt.figure(figsize=(5,5))
# 단위는 인치
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,4))
# 단위는 인치
plt.plot(x,y,'--g')
# 선 모양 색깔 순
plt.show()
```

```python
y = x**2
```

```python
plt.figure(figsize=(8,4))
# 단위는 인치
plt.plot(x,y,'--g')
# 선 모양 색깔 순
plt.show()
```

# 정규분포 그래프 그리기

```python
x = np.linspace(-5,5,100)
```

```python
sigma = 1.0
mean = 0.0
```

```python
np.pi
```

```python
y = (1.0/(np.sqrt(2.0*np.pi*sigma)))*np.exp(1.0)**-((x-mean)**2.0/(2.0*sigma**2.0))
```

```python
sigma2 = 2.0
```

```python
y2 = (1.0/(np.sqrt(2.0*np.pi*sigma2)))*np.exp(1.0)**-((x-mean)**2.0/(2.0*sigma2**2.0))
```

```python
sigma3 = 3.0
```

```python
y3 = (1.0/(np.sqrt(2.0*np.pi*sigma3)))*np.exp(1.0)**-((x-mean)**2.0/(2.0*sigma3**2.0))
```

```python
plt.plot(x,y,x,y2,x,y3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
